```python
# Import necessary libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 

from sklearn import linear_model
```

```python
# Create a dictionary data structure
data = {
    "Area": [2600, 3000, 3200, 3600, 4000],
    "Price": [5500000, 5650000, 6100000, 6800000, 7250000]
}

# Create a dataframe from the dictionary
df = pd.DataFrame(data, 
                  index = ["Katani", "Syokimau", "Athi River", "Kitengela", "Utawala"])

# Save the dataset as csv
df.to_csv("area_house.csv")
# Display the data
df
```

```python
# Visualise the data in a scatter plot to observe the patterns 
plt.figure(figsize = (10, 5))
plt.title("Relationship Between Area and House Price", fontsize = 14, fontname = 'serif')
plt.xlabel("Area (sq. ft)", fontsize = 12, fontname = 'serif')
plt.ylabel("House Price (KES in Million)", fontsize = 12, fontname = 'serif')
plt.scatter(df.Area, df.Price, color = "b", marker = "*")
```

```python
# Define the variables (feature and target)
# area = feature (independent variable)
print("Feature: Independent Varibale")
print("______________________________")
area = df.drop(columns = ["Price"])
print(area)
print("\n")

# price = target (dependent variable)
print(   "Target Variable")
print("_____________________")
price = df["Price"]
print(price)
```

```python
# Create a Simple Linear Regression Model 
reg = linear_model.LinearRegression()
```

```python
reg.fit(area, price)
```

```python
# intercept
b = reg.intercept_
b
```

```python
# Coefficient (Gradient)
m = reg.coef_
```

```python
predicted_price = reg.predict(df[["Area"]])
round(pd.DataFrame(predicted_price, columns = ["Predicted Price"], 
                   index = ["Katani", "Syokimau", "Athi River", "Kitengela", "Utawala"]), 2)
```

```python
# Visualise the Linear Regression Line (line of best fit)
plt.figure(figsize = (10, 5))
plt.title("Relationship Between Area and House Price", fontsize = 14, fontname = 'serif')
plt.xlabel("Area (sq. ft)", fontsize = 12, fontname = 'serif')
plt.ylabel("House Price (KES in Million)", fontsize = 12, fontname = 'serif')
plt.scatter(df.Area, df.Price, color = "b", marker = "*")
plt.plot(area, predicted_price, color = "g")
plt.show()
```
