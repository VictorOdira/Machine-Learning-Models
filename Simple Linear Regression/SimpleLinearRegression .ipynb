```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline 

from sklearn import linear_model
```

```python
df = pd.read_excel("Kenya_GDP _per_Capita.xlsx")
df.head()
```

```python
df.tail()
```

```python
year = df["Year"]
income = df["Income"]
```

```python
plt.figure(figsize = (10, 3))
plt.title("Kenya's Historical Per Capita Income", fontsize = 13, fontname = "serif")
plt.xlabel("Year", fontsize = 12, fontname = "serif")
plt.ylabel("Per Capita Income (US$)", fontsize = 12, fontname = "serif")
plt.scatter(year, income, color = "b", marker = "*")
plt.show()
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[["Year"]], df["Income"])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
predictions  = reg.predict(df[["Year"]])
predictions
```

```python
plt.figure(figsize = (10, 3))
plt.title("Kenya's Historical Per Capita Income", fontsize = 13, fontname = "serif")
plt.xlabel("Year", fontsize = 12, fontname = "serif")
plt.ylabel("Per Capita Income (US$)", fontsize = 12, fontname = "serif")
plt.scatter(year, income, color = "b", marker = "*")
plt.scatter(year, predictions, color = "r")
plt.plot(year, predictions, color = "g")
plt.show()
```
